```python
import matplotlib.pyplot as plt
```

```python
savings = [0,100,150,200,300,500,300,350,500,600,750,1000]
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
plt.plot(months,savings)
plt.xlabel('Month')
plt.ylabel('Savings ($)')
plt.title('Savings Account Results')
plt.show()
```

```python
# Create a Line Chart Using the MATLAB Method

%matplotlib inline

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create a plot using the object oriented method, for this we use ax.plot() instead of plt.plot()

# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Using plt.show()

# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Annotate Charts using MATLAB method
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()

# Add the legend.
plt.legend()
```

```python
# Annotate Charts Using the Object-Oriented Interface 
# (https://courses.bootcampspot.com/courses/95/pages/5-dot-1-4-annotate-charts?module_item_id=4145)

# Skill Drill

# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, 'go-', linewidth=2, markersize=5, marker='D')

# set line style
# ax.set_linestyle('-')

# Create labels for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

# Set the y limit between 0 and 45.
ax.set_ylim(0, 45)

# Create a title.
ax.set_title("PyBer Fare by Month")

# Add a grid.
ax.grid()

# Add the legend.
ax.legend('Boston')
```

```python
# Create a Scatter Plot Using the MATLAB Method
# (https://courses.bootcampspot.com/courses/95/pages/5-dot-1-7-create-scatter-plots-and-bubble-charts?module_item_id=4148)

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(x_axis, y_axis)
```

```python
"""Skill Drill
Using the Matplotlib MATLAB plotting approach, make the following changes to your scatter plot:
Change the color of the markers to red.
Add a legend for the city of Chicago.
Add a title and axes labels.
Switch the axis so that the Fare($) data is on the x-axis.
Add a limit for the x-axis data.
Invert the y-axis so that January is at the top"""
plt.scatter(y_axis, x_axis)
plt.gca().invert_yaxis()
plt.xlim(0, 50)

plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend('Chicago')
```

```python
# Create a Bubble Chart Using the Object-Oriented Interface

fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis*1000)
```

```python
"""Skill Drill
Using the object-oriented approach, make the following changes to your bubble chart:
Change the color of the markers to sky blue.
Change the size of the markers to 5 times each data point.
Make the color 20% transparent.
Add a black edge color to the circles.
Make the linewidth of the circles 2 points.
Add a legend for the city of Boston.
Add a title.
Switch the axis so that the Fare($) data is on the x-axis.
Add a limit for the x-axis data.
Invert the y-axis so that January is at the top.
"""
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=y_axis*5,c='skyblue',alpha=0.55,edgecolors='k',linewidths=2,)
ax.invert_yaxis()
ax.set_xlim(0, 50)
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")

ax.set_title("PyBer Fare by Month")


ax.legend('Chicago')
```

```python
# Create a Pie Chart Using the MATLAB Method

plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Create a Pie Chart Using the Object-Oriented Interface
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(y_axis,labels=x_axis)
plt.show()
```

```python
"""Skill Drill
Using the object-oriented approach, make the following changes to the pie chart:
Add a percentage to one decimal place to each wedge of the pie.
Increase the figure size to 8x8.
Explode the two highest percentage months.
Add a shadow.
Add a start angle so that January is at the top.
Reverse the order so that the month order is in a clockwise direction.
Add new colors of your own choosing or use the colors from the previous pie chart"""
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
fig, ax = plt.subplots(figsize=(8, 8))

ax.pie(y_axis,labels=x_axis,autopct='%.1f%%',explode=explode_values,startangle=-95,shadow='true',colors=colors)
ax.invert_yaxis()
plt.show()
```
